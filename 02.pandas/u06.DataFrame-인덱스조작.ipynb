```python
cctv = pd.read_csv('서울시CCTV설치운영현황(자치구)_년도별_210731기준.csv', skiprows = 1,  encoding='EUC-KR')
cctv.head()
```

```python
# 1)계 행을 삭제
cctv.drop([0], inplace=True)
cctv.head()
```

```python
# 2) 구분 열을 인덱스로 변경하고  이름을 자치구로 변경
cctv.set_index('구분',inplace=True)
cctv.index.name = '자치구'
cctv.head(3)
```

```python
# 3)인덱스에서 공백 삭제
cctv.index = cctv.index.str.replace(' ','')
cctv.head(3)
```

```python
# 4)데이터에서 천단위 구분기호를 지우고 정수로 변경
cctv = cctv.apply(lambda x : x.str.replace(',','')).astype(int)
cctv.head(3)
```

```python
# 5) 2017년 이전
cctv['2017년 이전'] = cctv['2012년 이전']+cctv['2012년']+cctv['2013년']+cctv['2014년']+cctv['2015년']+cctv['2016년']+cctv['2017년'] 
cctv.drop(columns=['2012년 이전','2012년','2013년','2014년','2015년','2016년','2017년','2021년'], inplace=True)
cctv.head(3)
```

```python
# 6)최근 3개월
cctv['최근3개년']=cctv['2018년']+cctv['2019년']+cctv['2020년']
cctv.drop(columns=['2018년','2019년','2020년'], inplace=True)
cctv.head(3)
```

```python
# 7)최근 증가율
cctv['최근증가율'] = (cctv['최근3개년']/cctv['2017년 이전']*100).round(2)
cctv.rename(columns = {'총계':'CCTV대수'}, inplace=True)
cctv.head(3)
```

```python
# 8) cctv 댓수 TOP5
cctv.sort_values(by='CCTV대수',ascending=False).head(5)
```

```python
# 9)최근 증가율 Top5
cctv.sort_values(by='최근증가율',ascending=False).head(5)
```

```python
# 10)csv로 저장
cctv.to_csv('cctv.csv')
```
