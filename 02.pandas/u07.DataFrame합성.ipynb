- 서울 인구 데이터와 cctv 데이터 합치기

```python
pop = pd.read_csv('인구.txt', sep= '\t', skiprows = 2)
pop.head()
```

```python
pop['총인구']=pop['계']
pop['노인인구']=pop['65세이상고령자']
pop = pop[['자치구','계','65세이상고령자']]
pop.head()
```

```python
pop.drop([0] ,inplace=True)
pop.head()
```

```python
pop.set_index('자치구', inplace=True)
pop.rename(columns={'계':'총인구','65세이상고령자':'노인인구'},inplace=True)
pop.head()
```

```python
pop = pop.apply(lambda x:x.str.replace(',','')).astype(int)
pop.head()
```

```python
pop['노인인구 비율'] = (pop.노인인구 / pop.총인구 * 100).round(2)
pop.head()
```

```python
cctv = pd.read_csv('cctv.csv')
cctv.head()
```

```python
cctv.set_index('자치구' , inplace=True)
cctv.head()
```

```python
df= pop.join(cctv)
df.head()
```

```python
df['1천명다cctv댓수'] = (df.CCTV대수/df.총인구 *1000).round(2)
df.tail()
```

```python
df.to_csv('서울시자치구별인구와CCTV.csv')
```
