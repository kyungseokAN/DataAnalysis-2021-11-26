```python
lines = []
naverin_url = f'https://kin.naver.com/'
driver = webdriver.Chrome('chromedriver.exe')
driver.get(naverin_url)
time.sleep(2)

req = requests.get(url)
html = req.text
for page in range(1,11):
    url = f'{search_url}{quote(keyword)}&page='+str(page)
    driver.get(url)
    time.sleep(1)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    lis = soup.select('.section>ul>li')

    for li in lis:
        href = li.find('a')['href']
        sub_url = href
        sub_req = requests.get(sub_url)
        sub_html = sub_req.text
        sub_soup = BeautifulSoup(sub_html,'html.parser')
        ans = sub_soup.select_one('._endContents.c-heading-answer__content > div._endContentsText.c-heading-answer__content-user').text
        title = sub_soup.select_one('.c-heading__title-inner>div:nth-last-child(1)').get_text().strip()
        
        lines.append([title,ans])
    
```

```python
df = pd.DataFrame(lines, columns=['제목','답변'])
```
