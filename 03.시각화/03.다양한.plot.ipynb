# matplotlib - 다양한 플롯

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'Malgun Gothic'
```

### 1.산점도(Scatter plot)

```python
import seaborn as sns
ans = sns.load_dataset('anscombe')
ans.head(3)
```

```python
ans.groupby('dataset').agg(['mean','std'])
```

```python
X,Y = np.random.randn(100),np.random.randn(100)
plt.title('산점도')
plt.scatter(X,Y)

plt.show()
```

```python
colors = np.random.rand(100)
area = np.pi*(15*np.random.rand(100))**2
plt.scatter(X,Y,s = area, c= colors, alpha = 0.5)
plt.title('Bubble chart')
plt.show()
```

### 2.Bar chart

```python
s = pd.Series(np.random.rand(10),index=list('abcdefghij'))
plt.bar(s.index, s.values)
plt.title('Bar chart')
plt.show()
```

```python
s.plot(kind='bar')
```

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic['class'].value_counts()
```

```python
s = titanic['class'].value_counts()
plt.bar(s.index, s.values)
plt.title('타이타닉호의 선실별 승객수')
plt.show()
```

```python
s = pd.Series(np.random.rand(10),index=list('abcdefghij'))
plt.barh(s.index,s.values)
plt.title('Barh chart')
plt.show()
```

### 3. Histogram

```python
s = np.random.randn(200)
plt.hist(s)
plt.title('Histogram')
plt.show()
```

```python
s = np.random.randn(200000)
plt.hist(s, bins=20)
plt.title('bins = 20개인 경우')
plt.show()
```

### 4. Box plot

```python
#Setosa 품종
iris = sns.load_dataset('iris')
setosa = iris[iris.species == 'setosa']
```

```python
plt.figure(figsize =(8,10) )
plt.boxplot([setosa.sepal_length, setosa.sepal_width,
             setosa.petal_length, setosa.petal_width],
             labels = iris.columns[:-1])
plt.title('Setosa 품종의 각 피쳐별 박스 플롯',fontsize = 15)
plt.show()
```

### 5.파이(pie) 차트

```python
ratio = [34,32,16,18]
labels = ['사과','바나나','멜론','포도']
plt.pie(ratio, labels=labels, autopct = '%.1f%%')
plt.show()
```

```python
labels = ['개구리', '돼지', '개', '통나무']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.title("Pie Chart")
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```
